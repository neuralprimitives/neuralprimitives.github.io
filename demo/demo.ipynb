## Set up VecUnico

Clone the repository and install dependencies. It may take 3-4 minutes.

```python
%%bash
git clone https://github.com/neuralprimitives/neuralprimitives.github.io.git
mv /content/neuralprimitives.github.io/demo /content/VecUnico 
cd VecUnico/extensions/pointnet2_ops_lib && python setup.py install
# install required PyPI packages
cd ../../build && pip install -r requirements.txt
# copy visual.py to root directory
cd ../..  && cp VecUnico/utils/visual.py visual.py
pip install -q condacolab
```

**Set up `abspy`.** We need `abspy` for polygon mesh reconstruction. The setup takes around **8 minutes (or slightly longer)**, since it requires building **SageMath** on Colab.

```python
import condacolab
condacolab.install()
```

```python
%%bash
cd VecUnico/compod
sudo apt-get update && sudo apt-get install libgomp1 ffmpeg libsm6 libxext6 -y

conda create -y --name compod python=3.10.10 && conda activate compod
conda install -y conda-forge::gsl anaconda::libgomp anaconda::scipy conda-forge::shapely conda-forge::sage conda-forge::tqdm conda-forge::trimesh conda-forge::treelib conda-forge::colorlog --verbose

pip install open3d gco-wrapper
```

##  Quick examples

```python
from visual import *
import subprocess

# -------------------------
# Config
# -------------------------
TEST_ID = "bag_0518100000337695"
PC_PATH  = f"VecUnico/evaluation/pc/{TEST_ID}.npy"
RESULT_PATH = f"VecUnico/evaluation/result/{TEST_ID}.obj"


def run(cmd: str):
    subprocess.run(cmd, shell=True, executable="/bin/bash", check=True)
```

**Example without transformation**

You can interactively explore the model by dragging it.

```python
# -------------------------
# 1) VecUnico inference
# -------------------------
run(f'cd /content/VecUnico/ && VECUNICO_TEST_ID="{TEST_ID}" /usr/bin/python3.real inference.py')

# -------------------------
# 2) CoMPoD demo (conda env)
# -------------------------
run(
    "source /usr/local/etc/profile.d/conda.sh "
    "conda activate compod &&"
    "cd VecUnico/compod &&"
    f"python example/single_vg_to_obj.py --input_file ../evaluation/vg/{TEST_ID}.vg --output_file ../evaluation/result/{TEST_ID}.obj && "
    "conda deactivate"
)

# -------------------------
# 3) Visualization
# -------------------------
visualize_point_cloud_and_mesh(
    "Incomplete point cloud", PC_PATH,
    "VecUnico reconstruction result", RESULT_PATH,
    eye=dict(x=0, y=1.4, z=0),
)
```

**Example with rotation transformation**

You can interactively explore the model by dragging it.

```python
# -------------------------
# 1) VecUnico inference
# -------------------------
run(f'cd /content/VecUnico/ &&  VECUNICO_TEST_ID="{TEST_ID}" VECUNICO_AUG_MODE="rotation" /usr/bin/python3.real inference.py')

# -------------------------
# 2) CoMPoD demo (conda env)
# -------------------------
run(
    "source /usr/local/etc/profile.d/conda.sh "
    "conda activate compod &&"
    "cd VecUnico/compod &&"
    f"python example/single_vg_to_obj.py --input_file ../evaluation/vg_rotation/{TEST_ID}.vg --output_file ../evaluation/result_rotation/{TEST_ID}.obj && "
    "conda deactivate"
)

# -------------------------
# 3) Visualization
# -------------------------
visualize_point_cloud_and_mesh(
    "Incomplete point cloud", PC_PATH.replace("pc/", "pc_rotation/"),
    "VecUnico reconstruction result", RESULT_PATH.replace("result/", "result_rotation/"),
    eye=dict(x=0, y=1.4, z=0),
)
```
